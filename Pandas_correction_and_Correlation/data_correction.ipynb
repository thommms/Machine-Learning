```python
#load the csv file into dataframe
df = pd.read_csv('housing-data.csv')[:-18]
df
```

```python
#checked for which column had null values and how many
df.isnull().sum()
```

```python
#show rows with null values
df[df.isnull().any(axis=1)]
```

```python
#replace all the zero values with NaN
df.replace(0,np.nan, inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
#gather all columns to replace NaN values except Car Garage column
cols = ['Suburb','Type','Price', 'Distance','Zipcode','#Bedroom','#Bathroom','Lot_size','Region_name','Property_count']
```

```python
#Replaced all the columns which has null values with median since they are sorted by suburb already except the '#-Car Garage' column because it's possible for a house not to have garage
df[cols] = df[cols].fillna(df[cols].median())
```

```python
df
```

```python
#we check if there is still any null value. So we can see that we have  just one in car-garage left. we can handle that later
df.isnull().sum()
```

```python
#replaced all the NaN with zero, because an apartment can have zero garage
df.replace(np.nan,0, inplace=True)
```

```python
df
```

```python
#Time to deal with negative impossible data (Bedroom,car Garage, Lot_size) e.g Bedroom can't have negative number of rooms
df[(df['#Bedroom']<0)]
```

```python
df[(df['#-Car Garage']<0)]
```

```python
df[(df['Lot_size']<0)]
```

```python
#here we handle the negative numbers in the 3 fields that contains negative values; Bedroom, Lot_size and Car Garage.
df['#Bedroom'] =df['#Bedroom'].mask(df['#Bedroom']<0,np.nan)
df['Lot_size'] =df['Lot_size'].mask(df['Lot_size']<0,np.nan)
df['#-Car Garage'] =df['#-Car Garage'].mask(df['#-Car Garage']<0,np.nan)
```

```python
df[(df['Property_count']==15510)]
```

```python
#we use the bfill method to replace the NaN. The bfill uses the next valid observation to fill the gap nearest
df['#Bedroom'].fillna(method='bfill',inplace = True)
df['Lot_size'].fillna(method='bfill',inplace = True)
df['#-Car Garage'].fillna(method='bfill',inplace = True)
```

```python
#checking the affected role to see if the NaN value has been reflected
df.loc[(df['Property_count']==15510) & (df['Lot_size']==630)]
```

```python
df.loc[(df['Property_count']==9264) & (df['Lot_size']==658)]
```
